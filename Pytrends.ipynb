```python
import pandas as pd                        
from pytrends.request import TrendReq

prov_list = ['BA','AC','BT','BE','YO','JK','GO','JA','JB','JT','JI','KB','KS','KT','KI','KU','BB','KR','LA','MA','MU','NB','NT','PA','PB','RI','SR','SN','ST','SG','SA','SB','SS','SU','ID']
prov_name = ['Bali','Aceh','Banten','Bengkulu','Daerah Istimewa Yogyakarta','Daerah Khusus Ibukota Jakarta','Gorontalo','Jambi','Jawa Barat','Jawa Tengah','Jawa Timur','Kalimantan Barat','Kalimantan Selatan','Kalimantan Tengah','Kalimantan Timur','Kalimantan Utara','Kepualuan Bangka Belitung','Kepulauan Riau','Lampung','Maluku','Maluku Utara','Nusa Tenggara Barat','Nusa Tenggara Timur','Papua','Papua Barat','Riau','Sulawesi Barat','Sulawesi Selatan','Sulawesi Tengah','Sulawesi Tenggara','Sulawesi Utara','Sumatera Barat','Sumatera Selatan','Sumatera Utara','Indonesia']

#Ganti timeframe sesuai waktu yang diinginkan
timeframe = '2021-01-07 2021-01-19'
#Ganti dengan keyword/topik yang diinginkan
keywords = ['gejala korona','masker','disinfektan','hand sanitizer','apd','corona','korona','covid']
for j in range(0,len(keywords)):
    print('Proses ',keywords[j])
    df = ''
    for i in range(0,len(prov_list)):
        print(prov_list[i])
        if(prov_list[i] == prov_list[0]):
            pytrend = TrendReq()
            pytrend.build_payload(kw_list=[keywords[j]],geo = 'ID-'+prov_list[i],timeframe = timeframe)
            df = pytrend.interest_over_time()
            if(len(df)==0):
                df[prov_name[i]] = 0
            else:
                df = df.drop(['isPartial'], axis=1)
                df['BA'] = df[keywords[j]]
                df = df.drop([keywords[j]], axis=1)
        elif(prov_list[i]=='ID'):
            pytrend = TrendReq()
            pytrend.build_payload(kw_list=[keywords[j]],geo = 'ID',timeframe = timeframe)
            df1 = pytrend.interest_over_time()
            if(len(df1)==0):
                df[prov_name[i]] = 0
            else:
                df[prov_name[i]] = df1[keywords[j]]
        else:
            pytrend = TrendReq()
            pytrend.build_payload(kw_list=[keywords[j]],geo = 'ID-'+prov_list[i],timeframe = '2021-01-07 2021-01-19')
            df1 = pytrend.interest_over_time()
            if(len(df1)==0):
                df[prov_name[i]] = 0
            else:
                df[prov_name[i]] = df1[keywords[j]]
    df.to_excel(keywords[j]+'.xlsx', index = True)
    print('Proses finish')
```
